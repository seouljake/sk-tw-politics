```python
url_stem = "https://zh.wikipedia.org/wiki/"
```

```python
for index,p in twtest.iterrows():
    
    # carry on where we last left off
    # later, let's write in an error catching loop to refresh the session and continue
    if(index>2038):

        # construct url
        testname = p["Name_Native"]
        url= url_stem+testname

        # request page
        x = requests.get(url)
        print(index,url,x.status_code)

        # status_code 200: article found
        # status_code 404: article not found

        # export page if status_code is 200
        if(x.status_code==200):
            fout = open(fileout_tw_path+"test_"+testname+".html","w",encoding="utf-8")
            fout.write(x.content.decode("utf-8"))
            fout.close()
```
