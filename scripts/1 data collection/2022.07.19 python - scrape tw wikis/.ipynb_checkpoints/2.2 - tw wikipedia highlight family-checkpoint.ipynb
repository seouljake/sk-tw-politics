### family terms

```python
terms_filename = "Family Member Search Terms.xlsx"
```

```python
terms_df = pd.read_excel(data_path + data_sub_path_2 + terms_filename)
terms_df
```

```python
terms_chinese_orig = terms_df.Chinese.to_list()
terms_chinese = []

while terms_chinese_orig:
    
    i = terms_chinese_orig.pop(0)
        
    ilist = i.split(";")
    
    terms_chinese += ilist
    
    # print(ilist,len(terms_chinese))

terms_chinese = [t.strip() for t in terms_chinese]
print(len(terms_chinese))
```

### wikipedia page list

```python
wiki_filename_list = os.listdir(data_path + data_sub_path_1)
len(wiki_filename_list)
```

# step 1. search tw wikipages for family terms

```python
filename1 = wiki_filename_list[0]
filename1
```

```python
page1 = open(data_path + data_sub_path_1 + filename1,encoding="utf-8")
```

```python
page1_text = page1.readlines()
```

```python
# prep code for a single page

page1_count = 0
page1_terms = []

for s in page1_text:
    for t in terms_chinese:
        if t in s:
            page1_terms.append(t)
            page1_count += 1

print(page1_count,page1_terms)
```

```python
# loop through all pages
family_metadat = pd.DataFrame(data = {"filename":[],"sentence_count":[],"term_count":[],"terms":[]},columns=["filename","sentence_count","term_count","terms"])

for f in wiki_filename_list:
    
    p = open(data_path + data_sub_path_1 + f,encoding="utf-8")
    p_text = p.readlines()
    
    t_count = 0
    t_list = []
    
    s_count = 0

    for s in p_text:
        
        s_true = False
        
        for t in terms_chinese:
            if t in s:
                t_list.append(t)
                t_count += 1
                s_true = True
                
        if s_true:
            s_count +=1
                
                
    t_str = ",".join(t_list)

    print(f,s_count,t_count,t_str)
    
    row = pd.DataFrame(data = {"filename":f,"sentence_count":s_count,"term_count":t_count,"terms":t_str},index=["filename"])
    
    family_metadat = pd.concat([family_metadat,row],ignore_index=True)
    
```
