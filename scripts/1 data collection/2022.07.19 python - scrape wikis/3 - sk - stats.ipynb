```python
import pandas as pd
import numpy as np
pd.set_option('display.max_rows', None)
pd.set_option('display.max_columns', None)
```

# import sk politician test data

```python
filein_path = "../../../data/data_wikis/1 - politician data for test scraping/"
```

```python
sktest = pd.read_excel(filein_path+"sk_testset.xlsx")
```

# website coverage of candidates & winners, overall

```python
# total
coverage0 = pd.DataFrame(index=["overall"])
coverage0["candidate_count"] = [sktest.shape[0]]
coverage0["candidate_withpage"] = [sktest[~(sktest.URL_NamuWiki==False)].shape[0]]
coverage0["candidate_coverage"] = coverage0["candidate_withpage"] / coverage0["candidate_count"]
coverage0["winner_count"] = [sktest[sktest.WonElection==1].shape[0]]
coverage0["winner_withpage"] = [sktest[~(sktest.URL_NamuWiki==False) & sktest.WonElection==1].shape[0]]
coverage0["winner_coverage"] = coverage0["winner_withpage"] / coverage0["winner_count"]
```

```python
coverage0
```

# website coverage of candidates & winners, by election type

```python
coverage = sktest.groupby(sktest.ElectionCategory_Native).agg(candidate_count=("Name_Native","count"))
```

```python
coverage["candidate_withpage"] = sktest[~(sktest.URL_NamuWiki==False)].groupby(sktest.ElectionCategory_Native).agg(candidate_coverage=("Name_Native","count"))
```

```python
coverage["candidate_coverage"] = coverage["candidate_withpage"] / coverage["candidate_count"]
```

```python
coverage["winner_count"] = sktest[sktest.WonElection==1].groupby(sktest.ElectionCategory_Native).agg({"Name_Native":"count"})
```

```python
coverage["winner_withpage"] = sktest[~(sktest.URL_NamuWiki==False) & sktest.WonElection==1].groupby(sktest.ElectionCategory_Native).agg({"Name_Native":"count"})
```

```python
coverage["winner_coverage"] = coverage["winner_withpage"] / coverage["winner_count"]
```

# coverage, by election type & overall
