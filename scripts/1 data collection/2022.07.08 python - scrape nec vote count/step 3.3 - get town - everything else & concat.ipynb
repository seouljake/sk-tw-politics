```python
import requests
from bs4 import BeautifulSoup
import pandas as pd
import json
```

# Step 3.3 - get town code - everything else & concatenate

```python
# import Step 2.3 - "get city" codes for everything else
filepath = "../../data/2.0 current through 2022 data/2022 NEC web scraping/step 2 - get city/dat_step2_3.xlsx"
dat_step2_3 = pd.read_excel(filepath,dtype=None)
```

```python
# "get town" url and post data for everything else
url = "http://info.nec.go.kr/bizcommon/selectbox/selectbox_townCodeBySgJson_Old.json"
electionId = "0000000000"
electionCode = ""   # same election code as in Step 2
subElectionCode = "" # same sub election code as in Step 2
cityCode = ""   # get city code from Step 2
dat_post = {'electionId': electionId, "subElectionCode": subElectionCode,"electionCode":electionCode,"cityCode":cityCode}
```

```python
dat_step3 = pd.DataFrame(columns=list(dat_step2_3.columns)+["TownCode","TownName"])
dat_step3
```

```python
dat_step2_3
```

```python
bodyparts=[]

for index,row in dat_step2_3.iterrows():
    dat_post["subElectionCode"]=row["SubElectionCode"]
    dat_post["electionCode"]=row["ElectionCode"]
    dat_post["cityCode"]=row["CityCode"]
    x = requests.post(url,data=dat_post)
    j = json.loads(x.text)
    body = j["jsonResult"]["body"]
    
    [bodyparts.append(list(row)+[item["CODE"],item["NAME"]]) for item in body]
    
annoy(500,500,6)
```

```python
bodyparts
```

```python
dat_step3_3 = pd.DataFrame(bodyparts,columns=dat_step3.columns)
dat_step3_3
```
