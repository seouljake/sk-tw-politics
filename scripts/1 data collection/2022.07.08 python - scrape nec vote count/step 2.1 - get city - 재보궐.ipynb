```python
import requests
from bs4 import BeautifulSoup
import pandas as pd
import json
```

```python
filepath = "../../data/2.0 current through 2022 data/2022 NEC web scraping/step 1 - get sub election/dat_step1.xlsx"
dat_step1 = pd.read_excel(filepath,dtype=None)
```

# Step 2.1 - get city code for 재보궐선거

```python
# Step 2 - city Code by Sg

# 재보궐선거
url = "http://info.nec.go.kr/bizcommon/selectbox/selectbox_cityCodeBySgJson_JAEBO_Old.json"
electionId = "0000000000"
subElectionCode = "" # get from the JSON response to Step 1
electionCode = ""   # same electionCode as in Step 1
dat_post = {'electionId': electionId, "subElectionCode": subElectionCode,"electionCode":electionCode}
```

```python
dat_step2 = pd.DataFrame(columns=list(dat_step1.columns)+["CityCode","CityName"])
dat_step2
```

```python
# 재보궐선거 from dat_step1  
dat_step1_subset = dat_step1[dat_step1["ElectionTypeCode"]==0]
dat_step1_subset.shape
```

```python
dat_step1_subset
```

```python
bodyparts=[]

for index,row in dat_step1_subset.iterrows():
    dat_post["subElectionCode"]=row["SubElectionCode"]
    dat_post["electionCode"]=row["ElectionCode"]
    x = requests.post(url,data=dat_post)
    j = json.loads(x.text)
    body = j["jsonResult"]["body"]
    
    [bodyparts.append(list(row)+[item["CODE"],item["NAME"]]) for item in body]
```

```python
bodyparts
```

```python
dat_step2_1 = pd.DataFrame(bodyparts,columns=dat_step2.columns)
dat_step2_1
```

```python
dat_step2_1.shape
```

```python
len(bodyparts)
```

```python
dat_step2_1.to_excel("../../data/2.0 current through 2022 data/2022 NEC web scraping/step 2 - get city/dat_step2_1.xlsx",index=False,encoding="utf-8")
```
