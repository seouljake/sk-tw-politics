```python
import winsound
frequency = 1000  # Set Frequency To 2500 Hertz
duration = 1000  # Set Duration To 1000 ms == 1 second

#winsound.Beep(frequency, duration)

def success(frequency,duration,n):
    for i in range(1, n):
        winsound.Beep(i * frequency, duration)
        
def fail(frequency,duration,n):
    for i in range(1,n):
        winsound.Beep((n-i) * frequency, duration)
        
success(500,500,6)
fail(500,500,6)
```

```python
# query sggTown - two urls - url_old for 지선 1,2,3;  url_new for everything else - no special url for 재보궐선거

url_new = "http://info.nec.go.kr/bizcommon/selectbox/selectbox_getSggTownCodeJson_Old.json"
url_old = "http://info.nec.go.kr/bizcommon/selectbox/selectbox_getSggTownCodeJson_GuOld.json"
electionId = "0000000000"
dat_post = {'electionId': electionId,
            "electionName":"",
            "electionCode":"",
            "townCode":""}
```

```python
try:
    
    bodyparts=[]

    for index,row in d.iterrows():
        #print(index)
        dat_post["electionName"]=row["ElectionCode"]
        dat_post["electionCode"]=row["SubElectionCode"]
        dat_post["townCode"]=row["TownCode"]

        # use url_old for 지선 1,2,3 and url_new for everything else
        if int(row["ElectionTypeCode"])==4 and int(row["ElectionCode"])<20060531:
            print(index,row["ElectionTypeCode"],row["ElectionCode"],"old")
            x = requests.post(url_old,data=dat_post)
        else:
            print(index,row["ElectionTypeCode"],row["ElectionCode"],"new")
            x = requests.post(url_new,data=dat_post)

        j = json.loads(x.text)
        body = j["jsonResult"]["body"]

        [bodyparts.append(list(row)+[item["CODE"],item["NAME"]]) for item in body]
        
    success(500,500,6)

except:
    
    fail(500,500,6)
    print("failed at",index)
```
