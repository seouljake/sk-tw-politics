```python
import requests
from bs4 import BeautifulSoup
import pandas as pd
import json
```

# Step 3.1 - get town code - 재보궐

```python
# import Step 2.1 - "get city" codes for 재보궐선거
filepath = "../../data/2.0 current through 2022 data/2022 NEC web scraping/step 2 - get city/dat_step2_1.xlsx"
dat_step2_1 = pd.read_excel(filepath,dtype=None)
```

```python
# "get town" url and post data for 재보궐선거 
url = "http://info.nec.go.kr/bizcommon/selectbox/selectbox_AllTownCodeBySgJson_JAEBO_Old.json"
electionId = "0000000000"
electionCode = ""   # same election code as in Step 2
subElectionCode = "" # same sub election code as in Step 2
cityCode = ""   # get city code from Step 2
dat_post = {'electionId': electionId, "subElectionCode": subElectionCode,"electionCode":electionCode,"cityCode":cityCode}
```

```python
dat_step3 = pd.DataFrame(columns=list(dat_step2_1.columns)+["TownCode","TownName"])
dat_step3
```

```python
dat_step2_1
```

```python
bodyparts=[]

for index,row in dat_step2_1.iterrows():
    dat_post["subElectionCode"]=row["SubElectionCode"]
    dat_post["electionCode"]=row["ElectionCode"]
    dat_post["cityCode"]=row["CityCode"]
    x = requests.post(url,data=dat_post)
    j = json.loads(x.text)
    body = j["jsonResult"]["body"]
    
    [bodyparts.append(list(row)+[item["CODE"],item["NAME"]]) for item in body]
```

```python
bodyparts
```

```python
dat_step3_1 = pd.DataFrame(bodyparts,columns=dat_step3.columns)
dat_step3_1
```
