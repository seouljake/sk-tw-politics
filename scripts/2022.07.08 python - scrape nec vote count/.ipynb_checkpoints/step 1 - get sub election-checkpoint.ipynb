```python
# Step 1 - get sub election type

url = "http://info.nec.go.kr/bizcommon/selectbox/selectbox_getSubElectionTypeJson.json"
electionId = "0000000000"
electionCode = ""
electionType = ""
dat_post = {'electionId': electionId, "electionCode": electionCode,"electionType":electionType}
```

```python
bodyparts=[]

for index,row in dat_step0.iterrows():
    dat_post["electionCode"]=row["ElectionCode"]
    dat_post["electionType"]=row["ElectionTypeCode"]
    x = requests.post(url,data=dat_post)
    j = json.loads(x.text)
    body = j["jsonResult"]["body"]
    
    [bodyparts.append(list(row)+[item["CODE"],item["NAME"]]) for item in body]
```

```python
dat_step1 = pd.DataFrame(bodyparts,columns=list(dat_step0.columns)+["SubElectionCode","SubElectionName"])
dat_step1
```
