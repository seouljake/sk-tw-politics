```python
import winsound
frequency = 1000  # Set Frequency To 2500 Hertz
duration = 1000  # Set Duration To 1000 ms == 1 second

#winsound.Beep(frequency, duration)

def success(frequency,duration,n):
    for i in range(1, n):
        winsound.Beep(i * frequency, duration)
        
def fail(frequency,duration,n):
    for i in range(1,n):
        winsound.Beep((n-i) * frequency, duration)
        
success(500,500,6)
fail(500,500,6)
```

```python
# query electionInfo page - for subelection = 1 (president)

url = "http://info.nec.go.kr/electioninfo/electionInfo_report.xhtml"

electionId = "0000000000"
requestURI = "/WEB-INF/jsp/electioninfo/0000000000/vc/vccp04.jsp"
topMenuId = "VC"
secondMenuId = "VCCP04"
menuId = "VCCP04"
statementId = "VCCP04_#2"
oldElectionType = "1"
electionType = "2"         # ElectionTypeCode - fixed - 총선 = 2, 
electionName = "20170509"  # ElectionCode - variable
electionCode = "7"         # SubElectionCode - fixed - 국회의원 비례대표 = 7
cityCode = "2600"          # CityCode - variable
sggCityCode = "-1"         # SggCityCode - NaN
townCodeFromSgg = "-1"     # SggTownCode - NaN
townCode = "2604"          # TownCode - variable
#sggTownCode = ""          # SGGTownCode - not a parameter
x = "62"
y = "20"

dat_post = {'electionId': electionId,
            "requestURI": requestURI,
            "topMenuId": topMenuId,
            "secondMenuId": secondMenuId,
            "menuId":menuId,
            "statementId":statementId,
            "oldElectionType":oldElectionType,
            "electionType":electionType,       # ElectionTypeCode - variable
            "electionName":electionName,       # ElectionCode - variable  
            "electionCode":electionCode,       # SubElectionCode - fixed
            "cityCode":cityCode,               # CityCode - variable
            "sggCityCode":sggCityCode,         # SggCityCode - fixed
            "townCodeFromSgg":townCodeFromSgg, # TownCode - fixed
            "townCode":townCode,               # TownCode - variable
            #"sggTownCode":"",                 # SggTownCode - omitted from subelection = 1 (president)
            "x":x,
            "y":y}
```

```python
excelpath = "../../data_electoral/2.0 current through 2022 data/2022 NEC web scraping/step 4 - get sggCity & sggTown/votecount_"+subelelection_name+"/"
```

```python
try:

    for index,row in d[660:1500].iterrows():

        # post parameters for subelection = 1 (president)
        dat_post["electionType"]=row["ElectionTypeCode"]
        dat_post["electionName"]=row["ElectionCode"]
        #dat_post["electionCode"]=row["SubElectionCode"]
        dat_post["cityCode"]=row["CityCode"]
        #dat_post["sggCityCode"]=row["SggCityCode"]
        #dat_post["townCodeFromSgg"]=row["TownCode"]
        dat_post["townCode"]=row["TownCode"]
        #dat_post["sggTownCode"]=row["SggTownCode"]
                
        x = requests.post(url,data=dat_post)
        #print(row)
        print(index,row["ElectionTypeName"],row["ElectionName"],row["ElectionCode"],row["CityCode"],row["CityName"],row["TownCode"],row["TownName"],x.status_code)

        #j = json.loads(x.text)

        # write html file
        # fo = open("out.html", "w",encoding="utf-8")
        # fo.write(x.text)
        # fo.close()


        # convert html table to pandas and write as an excel file
        tables = pd.read_html(x.text)
        t = tables[0]

        excelfilename = "votecount_"+subelelection_name+"_"+str(row["ElectionCode"])+"_"+str(row["CityCode"])+"_"+str(row["TownCode"])+".xlsx"

        t.to_excel(excelpath+excelfilename)

    success(500,500,6)
    
except:
    
    fail(500,500,6)
    
```
